```python
from pathlib import Path
from hloc.visualization import plot_images, plot_keypoints, read_image
from hloc.utils.io import get_keypoints
```

```python
DIR = "../image-matching-challenge-2023"
MODE = "train"
NAME = "SP+SG"
dataset = "heritage" # "heritage", "haiper", "urban"
scene = "cyprus" # "dioscuri", "cyprus", "wall", "kyiv-puppet-theater", "bike", "chairs", "fountain"
```

```python
scene_dir = Path(f"../outputs/{NAME}/{dataset}/{scene}")

images = Path(f"{DIR}/{MODE}/{dataset}/{scene}/images")
images_rotated = scene_dir / "images_rotated"

features = scene_dir / "features.h5"
features_rotated = scene_dir / "features_rotated.h5"
```

```python
image_fn = "DSC_6480.JPG"
```

### Rotated image and rotated keypoints

```python
plot_images([read_image(images_rotated / image_fn)])
plot_keypoints([get_keypoints(features_rotated, image_fn)], ps=2)
```

### Original image and "unrotated" keypoints

```python
plot_images([read_image(images / image_fn)])
plot_keypoints([get_keypoints(features, image_fn)], ps=2)
```
